# Steps

For age and date, remove values that are not in a certain business defined range

For price, remove negative values

For lat/lon, remove value that are not in the France box (see google map), assuming only France can be in base

# Sanitize feature by feature

```python
#exemple
def sanitize_age(dataframe, age_column_label='age', age_range=(18, 90)):
    age_to_keep_mask = dataframe[age_column_label].between(age_range[0], age_range[1])
    dataframe.loc[age_to_keep_mask == False, age_column_label] = np.nan
    return dataframe
```

```python
sanitize_age(loaded_fake_df)
```

```python
def sanitize_date(dataframe, date_column_label='date', date_range=(pd.to_datetime('1/1/1990'), pd.to_datetime('today'))):
    date_to_keep_mask = dataframe[date_column_label].between(date_range[0], date_range[1])
    dataframe.loc[date_to_keep_mask == False, date_column_label] = np.nan    
    return dataframe
```

```python
sanitize_date(loaded_fake_df)
```

```python
def sanitize_price(dataframe, price_column_label='price'):
    dataframe=dataframe[dataframe[price_column_label]>=0]
    return dataframe
```

```python
sanitize_price(loaded_fake_df)
```

```python
def sanitize_latitude(dataframe, latitude_column_label='latitude', latitude_range=(42, 54)):
    # From google maps, max point in Brussels, and min point in Barcelona (savy)
    dataframe=dataframe[(dataframe[latitude_column_label]>=42)&(dataframe[latitude_column_label]<=54)]
    return dataframe
```

```python
sanitize_latitude(loaded_fake_df)
```

```python
def sanitize_longitude(dataframe, longitude_column_label='longitude', longitude_range=(-4, 8)):
    # From google maps, max point in Stuttgard, and min point in Brest (savy)
    dataframe=dataframe[(dataframe[longitude_column_label]>=-4)&(dataframe[longitude_column_label]<=8)]
    
    return dataframe
```

```python
sanitize_longitude(loaded_fake_df)
```

# Wrap all in one function

```python
def sanitize_dataframe(dataframe):
    sanitize_age(dataframe)
    sanitize_date(dataframe)
    sanitize_price(dataframe)
    sanitize_latitude(dataframe)
    sanitize_longitude(dataframe)
    return dataframe
```

```python
sanitized_dataframe = sanitize_dataframe(loaded_fake_df)
```
