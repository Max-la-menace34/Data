```python
import pandas as pd
import plotly.express as px
import numpy as np
```

Step 2 : Context and data assessment

```python
climber_df = pd.read_csv('climber_df.csv')
climber_df['sex']=np.array(["man" if val==0 else "woman" for val in climber_df['sex']])
```

Questions:
- Are there differences bewteen men and women ?
- Are there differences bewteen the countries for a given sex ?
- Are old peaple experienced peaple ? Are they better ? Is there a correlation bewteen the grade and the experience ? 
- Are peaple improving their grade along the time ?
- What is the best morphology for climbing ? 


```python
climber_df_man=climber_df[climber_df["sex"]=="man"]
climber_df_woman = climber_df[climber_df["sex"]=="woman"]
```

```python
#- Are there differences bewteen men and women ?
fig = px.box(climber_df,x="sex",y="grades_mean")
fig.show()
```

```python
fig = px.violin(climber_df,x="sex",y="grades_mean")
fig.show()
```

As we can see, men have better grades than women. This can be explained by a difference in morphology, but also by the fact that the high level tracks are more in a massculin style (strength).

```python
#- Are old peaple experienced peaple ? Are they better ? Is there a correlation bewteen the grade and the experience ?
fig = px.histogram(climber_df_men,histfunc="avg",x="country",y="grades_mean")
fig.show()
```

As we can see some countries are better than others. For example  France, CZE (where the best climber of the world come from (Adam Ondra)) are one of the best.So we may find a correlation bewteen the country and the climbing level.

```python
#- Are old peaple experienced peaple ? Are they better ? Is there a correlation bewteen the grade and the experience ?

fig = px.scatter(climber_df, x="age", y="grades_mean")
fig.show()
```

As we can see the level is decreasing along the time,but I am surprised that the difference is not even greater. Indeed, climbing is a very rude sport so young people have a better capacity to solve this problem.  

```python
#- Are there differences bewteen the countries for a given sex ?
#- Are old peaple experienced peaple ? Are they better ? Is there a correlation bewteen the grade and the experience ? 

fig = px.scatter(climber_df, x="grades_mean", y="age",
	         size="grades_mean", color="sex",log_x=True, size_max=20,title="Buble plot")
fig.show()
```

As we can see most the best grades are made by young men , this way of representing is the same as the previous graph but just in one.

```python
#- What is the best morphology for climbing ? 
fig = px.histogram(climber_df,histfunc="avg", y="weight",x="height",  facet_col="sex",color="sex")
fig.show()
```

Most of the climbers are tall and not very heavy. But we don't really know about what is the best morpho.

```python
fig = px.scatter(climber_df, x="height", y="weight",
	         size="grades_max", color="grades_mean",log_x=True, size_max=30,title="Buble plot",width=1600, height=1000)

fig.show()
```

As we can see there is no particular morphology. However, is there a correlation with the BMI ?

```python
fig = px.scatter(climber_df[climber_df["grades_max"]>65], x="height", y="weight",
	         size="grades_max", color="grades_mean",log_x=True, size_max=30,title="Buble plot",width=1600, height=1000)

fig.show()
```

```python
climber_df["imc"]=round(climber_df["weight"]/(climber_df["height"]*climber_df["height"]*0.0001),0)
climber_df_imc=climber_df[climber_df["imc"].between(15,30)]
```

```python
climber_df.head()
```

```python
fig = px.box(climber_df_imc,x="imc",y="grades_max")
fig.show()
```

We notice that the BMI between 18 and 22 have better results. So we can confirm that there is a correlation between the morphology and the climbing level.
