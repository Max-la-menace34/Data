```python
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
```

```python
climber_df = pd.read_csv('climber_df.csv')
```

```python
climber_df.head()
```

```python

#climber_df['age']=np.array(["man" if val==0 else "woman" for val in climber_df['sex']])


climber_df["imc"]=round(climber_df["weight"]/(climber_df["height"]*climber_df["height"]*0.0001),0)
climber_df_imc=climber_df[climber_df["imc"].between(15,30)]
dictionnaire = ["height","weight","age","grades_count","grades_mean","grades_max","imc"]


for index in dictionnaire:
   climber_df[index]=((climber_df[index]-climber_df[index].min())/(climber_df[index].max()-climber_df[index].min()))*5


```

```python
#Split by age 
age_cut= climber_df["age"].median()
climber_young = climber_df[climber_df["age"]<=age_cut]
climber_old = climber_df[climber_df["age"]>age_cut]
```

```python
# Split by sex 
climber_man = climber_df[climber_df["sex"]==0]
climber_woman = climber_df[climber_df["sex"]==1]
```

```python
climber_old.head()
```

```python


categories = ['height','weight',
              'grades_max',"imc"]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[climber_old["height"].mean(),climber_old["weight"].mean(),climber_old["grades_max"].mean(),climber_young["imc"].mean()],
      theta=categories,
      fillcolor='cyan',
      opacity = 0.5,
      fill='toself',
      name='Old peaple more than 30 years old '
))
fig.add_trace(go.Scatterpolar(
      r=[climber_young["height"].mean(),climber_young["weight"].mean(),climber_young["grades_max"].mean(),climber_young["imc"].mean()],
      theta=categories,
      fillcolor='pink',
      opacity = 0.7,
      fill='toself',
      name='Young peaple'
))


fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,    
      range=[0, 4]
    )),
  showlegend=True
)

fig.show()
```

```python
categories = ['height','Weight',
              'grades_max',"BMI","Age"]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[climber_man["height"].mean(),climber_man["weight"].mean(),climber_man["grades_max"].mean(),climber_man["imc"].mean(),climber_man['age']],
      theta=categories,
      fillcolor='cyan',
      opacity = 0.5,
      fill='toself',
      name='Man'
))
fig.add_trace(go.Scatterpolar(
      r=[climber_woman["height"].mean(),climber_woman["weight"].mean(),climber_woman["grades_max"].mean(),climber_woman["imc"].mean(),climber_woman['age']],
      theta=categories,
      fillcolor='pink',
      opacity = 0.7,
      fill='toself',
      name='Woman'
))


fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,    
      range=[0, 4]
    )),
  showlegend=True
)

fig.show()
```
