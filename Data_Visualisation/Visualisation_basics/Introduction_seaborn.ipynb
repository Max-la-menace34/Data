Let's try to use this overview of variables for **all the numerical variables present** (to see their distribution)

Tips: see the pandas fonction [select_dtypes](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.select_dtypes.html)

It returns a part of the dataframe, and we want to keep only the columns, you can use `dataframe.columns` to get them as a list

```python
# Create distribution plots for all numerical variables
# - First, get numerical values
df_num=df_2020.select_dtypes(include="float64")
print(df_num.shape)

```

- As we want to create many plots of the same kind, we will need [matplotlib subplots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) -> `fig, ax = plt.subplots(X, Y, figsize=(A, B))`

- Than we will need to iterate over both **numerical variables** and **subplots** and for this we can use the python iterator function [zip](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) -> `for variable, subplot in zip(VAR, ax.flatten()):`


- Finally we can create a for loop to create a chart for each of them. In that case, instead of **distplot** we will use **histoplot** which can manage axes settings -> `sns.COMMAND(df[variable], ax=subplot)`

```python
# - Then create the plots
fig , ax = plt.subplots(2,4, figsize=(10, 10))
for variable, subplot in zip(df_num.columns, ax.flatten()):
    sns.histplot(df_num[variable], ax=subplot)
```

What kind of plot can we use to see the relationship between one numerical variable and two categorical variables ? 

For example, how can you visualize the happiness score, the gdp per capita and the region variables ? Make at least one chart.


```python
# 1 Categorical x 2 Numerical : Chart for the relationship between 2 numerical & 1 categorical variable
# (happiness score, the gdp per capita and the region variables)
sns.set_theme(style="white", color_codes=True)

# Use JointGrid directly to draw a custom plot
g = sns.JointGrid(data=df_2020, x="score", y="freedom_life_choices", space=0, ratio=17)
g.plot_joint(sns.scatterplot, size=df_2020["region"], sizes=(30, 120),
             color="g", alpha=.6, legend=False)
g.plot_marginals(sns.rugplot, height=1, color="g", alpha=.6)

sns.set_theme(style="whitegrid")


# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="score", y="freedom_life_choices",size="region",sizes=(2, 10), linewidth=0,data=df_2020, ax=ax)
ax.legend(bbox_to_anchor=(1, 0.6),loc='lower left')




```

If its not lisible, how could you plot small multiples, e.g. one subplot for each region (the categorical variable) ?

- Tips / small multiples : [FacetGrid](https://seaborn.pydata.org/tutorial/axis_grids.html#conditional-small-multiples)
- Tips / Plot size : col_wrap=VAR, height=VAR

```python
# How could you plot small multiples, e.g. one subplot for each region ?
g = sns.FacetGrid(df_2020, col="region",col_wrap=3)
g.map(sns.scatterplot, "freedom_life_choices", "score", alpha=1)
g.add_legend()
```

### B) Looking into the past

Here we want to explore further about the evolution of the happiness scores over time.

In the notebook template, I already merged data from original files between 2015 and 2020. 


```python
# Importing data from other files, adding a 'year' column
df_2019 = pd.read_csv("data/2019.csv")[['country', 'score']]
df_2019['year'] = 2019
df_2018 = pd.read_csv("data/2018.csv")[['country', 'score']]
df_2018['year'] = 2018
df_2017 = pd.read_csv("data/2017.csv")[['country', 'score']]
df_2017['year'] = 2017
df_2016 = pd.read_csv("data/2016.csv")[['country', 'score']]
df_2016['year'] = 2016
df_2015 = pd.read_csv("data/2015.csv")[['country', 'score']]
df_2015['year'] = 2015

df_2020['year'] = 2020
df_2020_score = df_2020[['country', 'score', 'year']]
```

```python
# Merging all files together. 
# In the seaborn world, this shape is called Long- form : https://seaborn.pydata.org/tutorial/data_structure.html
df_merge = pd.concat([df_2020_score, df_2019, df_2018, df_2017, df_2016, df_2015]).reset_index()
df_merge.head()
```

Lets see the evolution of happiness scores for one country over time.

What type of charts can show the evolution relationship of a variable ? 

```python
# Evolution of happiness for one country

# - Filter the dataframe on one country (select specific rows)
#   Tips : Tips / select rows based on values with pandas : df_filtered = df.loc[condition]
df_filtered = df_merge[df_merge["country"]=="Norway"]
```

```python
# - Plot the chart
sns.scatterplot(x="year",y="score",data=df_filtered)
sns.lineplot(x="year",y="score",data=df_filtered)
```

Lets plot the evolution for **all countries**. 

```python
# Lets plot the evolution for all countries. Is that lisible ?
plt.figure(figsize=(20,8))
sns.lineplot(
    data=df_merge,
    x="year", y="score", units="country",
    estimator=None, lw=1,
)
```

Lets try to look at this with **2 different approaches** :

**1/** Separating countries that have a **positive / negative trend**.

We are defining here a new variable to add meaning : trend, which is True if the happiness score in 2020 is higher than the score in 2015. (This is a limited proxy for assessing a trend, but well use it because its simple enough).

```python
# Now I want to distinguish countries where the trend is positive (2020 score > 2015 score) versus negative
# For this, I need to compute the trend value by country. We will use the pandas'pivot functions
# to easily compute a new variable for each country
# In the seaborn world, this shape is called Wide-form : https://seaborn.pydata.org/tutorial/data_structure.html
pivot = df_merge.pivot(index ='country', columns='year', values='score')
pivot.head()
```

```python
# computing the trend value for each country
df_pivot = pd.DataFrame(pivot.to_records())
df_pivot['trend'] = (df_pivot['2020'] - df_pivot['2015']) > 0
df_pivot.head()
```

```python
# now we want to go back to the Long-form format, it's often easier to plot in seaborn
df_melted = pd.melt(df_pivot, id_vars=['country', 'trend'], value_vars=['2015', '2016', '2017', '2018', '2019', '2020'], var_name='year', value_name='score')
df_melted.head()
```

```python
# plotting the same chart with two tweaks : we are separating the line both physically and by color coding
# according to the value of the variable Trend
sns.relplot(
    data=df_melted,
    x="year", y="score",
    hue="trend", col="trend",
    units='country', estimator=None,
    kind="line",
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

*2/* Looking at more specific groups of countries : the **most and the least happy**. The process on the most happy countries is provided. 

```python
# Let's arbitrarily define "happy countries" by the fact of having a score higher than 7 in 2020
df_high_happiness = df_merge[(df_merge['year'] == 2020) & (df_merge['score'] > 7)]
df_high_happiness.head()
```

```python
# we extract those countries's name
happy_countries = pd.Series(df_high_happiness['country'].unique())
happy_countries
```

```python
# back to our data, we are creating a new dataframe filtering th original one with those selected countries's name
df_happy_countries = df_merge.loc[df_merge['country'].isin(happy_countries.values)]
```

```python
# in order to see the evolution clearly for each country, we are using a small multiple plot
# note how the x and y axis have the same scale so that it is easier to compare visualy all the countries
# we are using a linked point plot here to keep in mind that it's an evolution, but we don't have really linear data
g = sns.FacetGrid(df_happy_countries, col="country", col_wrap=5, height=3)
g.map(sns.pointplot, "year", "score", order=[2015, 2016, 2017, 2018, 2019, 2020])
```
